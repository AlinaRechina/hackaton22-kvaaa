```python
import pandas as pd
df = pd.read_csv("wiki_movie_plots_deduped.csv", sep=",")
df
```

```python
d = df.groupby("Release Year").agg({"Title": list})
d
```

```python
films = {}
for i in range (1901, 2018):
    films[i] = df[df["Release Year"] == i]["Title"]
```

```python
lengths = {}
for year in films.keys():
    names = list(films[year])
    c = 0
    summ = 0
    for k in names:
        c += 1
        length = len(k.split(" "))
        summ += length
        aver = summ/c
    lengths[year] = round(aver, 2)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X_years = list(lengths.keys())
Y_names = list(lengths.values())
```

```python
plt.figure(figsize= (15, 7))
plt.bar(X_years, Y_names, color='pink')
plt.title('Среднее количество слов в названии фильмов разных лет')
plt.ylabel('Среднее количество слов')
plt.xlabel('Год')
plt.show()
```
