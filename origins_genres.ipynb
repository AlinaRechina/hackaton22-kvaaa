```python
import pandas as pd
df = pd.read_csv("wiki_movie_plots_deduped.csv")
df
```

```python
origins = []
v = df["Origin/Ethnicity"].sort_values()
for i in v:
    if i not in origins:
        origins.append(i)
```

```python
origins_genres = {}
for ori in origins:
    origins_genres[ori] = df[df["Origin/Ethnicity"] == ori]["Genre"]
```

```python
d = {}
for i in origins_genres.keys():
    counter = 0
    l = list(origins_genres[i])
    word = l[0]
    for j in l:
        curr_frequency = l.count(j)
        if((curr_frequency > counter) & (j != "unknown")):
            counter = curr_frequency
            word = j
    d[i] = word
```

```python
!pip install tabulate
from tabulate import tabulate
print(tabulate(d.items(), headers=['Origin/Ethnicity', 'Genre'], tablefmt = "pretty", stralign='center'))
```
